# Analysis of Results

```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def load_training_history(name):
    return pd.read_csv(os.path.join(name, 'training_log.csv'))

model = load_training_history('lstm100')

model['perplexity'] = np.exp(model['loss'])
model['val_perplexity'] = np.exp(model['val_loss'])

model.head(10)
```

```python
model.sort_values('val_perplexity', ascending=True).head(10)
```

```python
plt.style.use('fivethirtyeight')

fig, ax1 = plt.subplots(figsize=(8,5))

color = 'tab:red'
ax1.plot(model['epoch'], model['loss'], color=color)
ax1.plot(model['epoch'], model['val_loss'], 'r--', color=color)
ax1.set_ylabel('cross-entropy loss', color=color)
ax1.set_xlabel('epoch')
# ax1.legend(['train', 'val'])

ax2 = ax1.twinx() 
color = 'tab:green'
ax2.plot(model['epoch'], model['top10_acc'], color=color)
ax2.plot(model['epoch'], model['val_top10_acc'], 'g--', color=color)
# ax2.legend(['train', 'val'])
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylabel('top10 accuracy', color=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title("Simple LSTM")
plt.show()
```

```python
model2 = load_training_history('lstm100_artist100_hidden100')
model2.head()
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))

ax1.plot(model['epoch'], model['val_perplexity'])
ax1.plot(model2['epoch'], model2['val_perplexity'])
ax1.set_ylabel('perplexity')
ax1.set_xlabel('epoch')
ax1.set_title("perplexity")
ax1.legend(["baseline", "+ artist"], fontsize=14)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))

ax1.plot(model['epoch'], model['val_top10_acc'])
ax1.plot(model2['epoch'], model2['val_top10_acc'])
ax1.set_ylabel('top10 accuracy')
ax1.set_xlabel('epoch')
ax1.legend(["baseline", "+ artist"], fontsize=14)
ax1.set_title("top10 accuracy")
plt.show()
```
